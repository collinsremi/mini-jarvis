# STEP 2: Prepare Feature Extractor and Prepocessing

```python
from transformers import WhisperFeatureExtractor
from transformers import WhisperTokenizer
```

```python
feature_extracor =  WhisperFeatureExtractor.from_pretrained("openai/whisper-small")

tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-small", language = "ja", task = "transcribe")
```

# STEP 4: Prepare Data

```python
from datasets import Audio

print("checking random samples of our common voice audio")
print(f"{common_voice["Train"][0]}/n")
```

```python
common_voice = common_voice.cast_column("audio", Audio(sampling_rate = 16000))

print("checking random samples of our common voice audio after down sampling from 42khz to 16khz")
print(f"{common_voice["Train"][0]}\n")
```

```python
def prepare_datasets(batch):

    audio = batch["audio"]

    batch["input_features"] = feature_extracor(audio["array"], sampling_rate = audio["sampling_rate"]).input_features[0]

    batch["labels"] = tokenizer(batch["sentence"]).input_ids

    return batch
```
